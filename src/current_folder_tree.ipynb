```python
# %%
from pathlib import Path



ROOT = Path("/Users/y-osamu/study/poi_sim")

EXCLUDE_DIRS = {
    ".git",
    "__pycache__",
    ".ipynb_checkpoints",
    ".venv",
    "node_modules",
}
EXCLUDE_FILES = {
    ".DS_Store",
}

def is_under_data_dir(path: Path) -> bool:
    """data 配下かどうか"""
    try:
        path.relative_to(ROOT / "data")
        return True
    except ValueError:
        return False

def walk(dir_path: Path, prefix: str = ""):
    items = []
    for child in dir_path.iterdir():
        # 除外ディレクトリ
        if child.is_dir() and child.name in EXCLUDE_DIRS:
            continue

        # data 配下では「ディレクトリのみ」表示
        if is_under_data_dir(child):
            if not child.is_dir():
                continue
        else:
            # data 以外では不要ファイルを除外
            if child.is_file() and child.name in EXCLUDE_FILES:
                continue

        items.append(child)

    # ディレクトリ → ファイルの順で安定ソート
    items.sort(key=lambda p: (not p.is_dir(), p.name.lower()))

    for i, p in enumerate(items):
        is_last = (i == len(items) - 1)
        branch = "└── " if is_last else "├── "
        print(prefix + branch + p.name + ("/" if p.is_dir() else ""))
        if p.is_dir():
            extension = "    " if is_last else "│   "
            walk(p, prefix + extension)

print(ROOT.name + "/")
walk(ROOT)
```
